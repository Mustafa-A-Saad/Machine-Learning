```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
digits['data'][0] # same as images but data is numerical
```

```python
plt.gray()
plt.matshow(digits.images[0]) # image is image
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(digits['data'],digits['target'],test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(max_iter=10000).fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(X_test)
```

```python
# We using Confusion Matrix to see where the error happen
```

```python
y_predicted = clf.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_predicted)
```

```python
# Confusion Matrix Visualize
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.show()
```

```python
from sklearn.datasets import load_iris
```

```python
df = load_iris()
dir(df)
```

```python
df['feature_names']
```

```python
df['data'][0] 
```

```python
df['target_names']
```

```python
df['target'] 
```

```python
X = df['data']
y = df['target']
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.3)
```

```python
clf = LogisticRegression(max_iter=1000).fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
clf.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy of the model:", accuracy)
```

```python
clf.score(X_test,y_test)
```

```python
sample_to_predict = X_test[:5]
predicted_class = clf.predict(sample_to_predict)
for i,prediction in enumerate(predicted_class):
    print(f"Sample {i+1}: Predicted class - {df['target_names'][prediction]}")
```
