```python
import pandas as pd
df = pd.read_csv("Excels\\titanic.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(['Name','PassengerId','Embarked','Cabin','Ticket','Parch','SibSp'],axis='columns')
df.head()
```

```python
dummies = pd.get_dummies(df['Sex'])
dummies.head()
```

```python
finaldf = pd.concat([df,dummies],axis='columns')
```

```python
finaldf = finaldf.drop(['Sex','female'],axis='columns')
```

```python
finaldf.head()
```

```python
finaldf['Age'] = finaldf['Age'].fillna(finaldf['Age'].median())
```

```python
X = finaldf.drop('Survived',axis='columns')
y = finaldf['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
X_train
```

```python
model.fit(X_train,y_train)
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test[:10])
```
