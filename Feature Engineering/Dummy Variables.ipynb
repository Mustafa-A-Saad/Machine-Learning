```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
```

```python
df = pd.read_csv("Excels\\homeprices3.csv")
df.head()
```

```python
dummies = pd.get_dummies(df['town'])  # One-Hot-Encoding for Nomial Catergoical variables
dummies 
```

```python
mergedf = pd.concat([df,dummies],axis='columns')
mergedf
```

```python
# to avoid dummy variables trap we remove one of the dummies colums and we don't need the town col anymore
```

```python
finaldf = mergedf.drop(['town','west windsor'],axis="columns")
finaldf
```

```python
# X = finaldf[['area','monroe township','robinsville']] # same as below
X = finaldf.drop(['price'],axis='columns')
y = finaldf['price']
reg = linear_model.LinearRegression().fit(X,y)
```

```python
X
```

```python
reg.predict([[2800,0,1]]) # robinsville
```

```python
reg.predict([[3400,0,0]]) # west winsdor predict
```

```python
reg.score(X,y) # 95% accurate
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
dfle = df
dfle['town'] = le.fit_transform(dfle.town) # takes label column as input and switch to numerical 
```

```python
dfle
```

```python
X = dfle[['town','area']].values  # .values gives back a 2d Array
y = dfle['price'].values
X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
reg.fit(X,y)
```

```python
reg.predict([[1,0,2800]])
```

```python
reg.score(X,y)
```

```python
df = pd.read_csv("Excels\\carprices.csv")
df
```

```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
mergedf = pd.concat([df,dummies],axis='columns')
mergedf
```

```python
mergedf.columns
```

```python
mergedf = mergedf.rename(columns = {
    'Car Model_BMW X5' : 'BMW X5',
    'Car Model_Audi A5' : 'Audi A5',
    'Car Model_Mercedez Benz C class' : 'Mercedez Benz C',
    "Sell Price($)" : "price"
})
```

```python
finaldf = mergedf.drop(['Car Model','Audi A5'],axis='columns')
finaldf
```

```python
X = finaldf.drop(['price'],axis='columns')
X
```

```python
y = finaldf['price']
y
```

```python
reg = linear_model.LinearRegression().fit(X,y)
```

```python
reg.predict([[45000,4,0,1]])
```

```python
reg.predict([[86000,7,1,0]])
```

```python
reg.score(X,y)
```
