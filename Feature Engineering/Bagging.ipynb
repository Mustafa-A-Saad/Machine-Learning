```python
df = pd.read_csv("Excels\\heart.csv")
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df['z-score'] = (df['Cholesterol'] - df['Cholesterol'].mean())/df['Cholesterol'].std()
df.head()
```

```python
df[ (df['z-score'] > 3) | (df['z-score'] < -3) ]
```

```python
df = df[ (df['z-score'] < 3) & (df['z-score'] > -3) ]
df.head()
```

```python
df[df['Oldpeak'] > df['Oldpeak'].mean()+3*df['Oldpeak'].std()]
```

```python
df = df[df['Oldpeak'] <= df['Oldpeak'].mean()+3*df['Oldpeak'].std()]
```

```python
df[df.RestingBP>(df.RestingBP.mean()+3*df.RestingBP.std())]
```

```python
df = df[df.RestingBP<=(df.RestingBP.mean()+3*df.RestingBP.std())]
```

```python
df.shape
```

```python
df.ExerciseAngina.replace(
    {
        'N': 0,
        'Y': 1
    },
    inplace=True)

df.ST_Slope.replace(
    {
        'Down': 1,
        'Flat': 2,
        'Up': 3
    },
    inplace=True
)

df.RestingECG.replace(
    {
        'Normal': 1,
        'ST': 2,
        'LVH': 3
    },
    inplace=True)

df.head()
```

```python
dummies = pd.get_dummies(df,drop_first=True)
dummies
```

```python
df = dummies
df
```

```python
df = df.drop(['z-score'],axis='columns')
```

```python
df.head()
```

```python
X = df.drop(['HeartDisease'],axis='columns')
y = df['HeartDisease']
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC

X_train , X_test , y_train , y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=10)
model = SVC().fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
scores = cross_val_score(SVC(),X,y,cv=5)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier

bag_model = BaggingClassifier(
        base_estimator=SVC(), 
        n_estimators=100, 
        max_samples=0.8, 
        random_state=0
        )

bagging_model.fit(X_train,y_train)
bagging_model.score(X_test,y_test)
```

```python
scores = cross_val_score(bag_model,X,y,cv=10)
scores.mean() # Bagging model works best with models with high variance and unstable , here SVC without bag is better
              # try with DecisionTreeClassifier
```

```python
from sklearn.tree import DecisionTreeClassifier

scores = cross_val_score(DecisionTreeClassifier(random_state=0),X,y,cv=5)
scores.mean() 
```

```python
bag_model = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(random_state=0), 
    n_estimators=100, 
    max_samples=0.9, 
    oob_score=True,
    random_state=0
)

scores = cross_val_score(bag_model, X, y, cv=5)
scores.mean() # 
```
