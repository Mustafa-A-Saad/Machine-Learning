```python
import pandas as pd
df = pd.read_csv("Excels\\spam.csv")
df.head()
```

```python
df.groupby("Category").describe()
```

```python
dummies = pd.get_dummies(df['Category'])
```

```python
df = pd.concat([df,dummies],axis='columns')
df.head()
```

```python
df.drop(['Category','ham'],axis='columns')
```

```python
from sklearn.model_selection import train_test_split
X = df['Message']
y = df['spam']
X_train, X_test , y_train , y_test = train_test_split(X,y,test_size=0.25)
```

```python
# Very good for Message Features or Text Features ( email spam )
from sklearn.feature_extraction.text import CountVectorizer
vc = CountVectorizer()
X_train_count = vc.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = vc.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = vc.transform(X_test)
model.score(X_test_count,y_test)
```

#-----------Here a better way with a Pipeline

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer',CountVectorizer()),
    ('nb',MultinomialNB())
])
```

```python
clf.fit(X_train,y_train) # after pipe we can just fit it
```

```python
clf.predict(X_test)
```

```python
clf.score(X_test,y_test)
```

```python
clf.predict(emails)
```
