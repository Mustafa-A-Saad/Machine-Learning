```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import linear_model
```

```python
df = pd.read_csv("Excels\\homeprices.csv")
df
```

```python
plt.scatter(x=df.area,y=df.price,c='red',marker="+")
plt.xlabel("Area (Sqrt Ft)")
plt.ylabel("Price (USD)")
```

```python
X = df[['area']] # df[['area']] here means that we only take the area column and drop others # X (Target/features take 2d arrays)
y = df['price'] 
reg = linear_model.LinearRegression().fit(X,y) 
reg.predict([[3300]])
```

```python
reg.coef_ # this is the slope of the LinearRegression
```

```python
reg.intercept_ # this is the y-intercept
```

```python
plt.scatter(x=df.area,y=df.price,c='red',marker="+")
plt.plot(df.area,reg.predict(df[['area']])) # LinearRegression Plot
plt.xlabel("Area (Sqrt Ft)")
plt.ylabel("Price (USD)")
```
