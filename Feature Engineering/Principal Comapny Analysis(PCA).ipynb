```python
df = pd.read_csv("Excels\\heart.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
_,mean,std,*_ = df['Cholesterol'].describe()
```

```python
mean,std
```

```python
df['z-score'] = (df['Cholesterol']-df['Cholesterol'].mean())/df['Cholesterol'].std()
df.head()
```

```python
df[(df['z-score']>3) | (df['z-score'] < -3)] # outliers
```

```python
df = df[(df['z-score']<3) & (df['z-score']>-3)]
```

```python
df[df['MaxHR']>df['MaxHR'].mean()+3*df['MaxHR'].std()]
```

```python
df[df['Oldpeak']>df['Oldpeak'].mean()+3*df['Oldpeak'].std()] # outliers
```

```python
df = df[df['Oldpeak']<=df['Oldpeak'].mean()+3*df['Oldpeak'].std()]
df.shape
```

```python
df[df['RestingBP']>df['RestingBP'].mean()+3*df['RestingBP'].std()]
df.shape
```

```python
df2 = pd.get_dummies(df,drop_first=True)
df2.head()
```

```python
dummies1 = pd.get_dummies(df['Sex'],prefix='Gender')
dummies2 = pd.get_dummies(df['ChestPainType'],prefix='ChestPain')
dummies3 = pd.get_dummies(df['ExerciseAngina'],prefix='ExerciseAng')
df['ST_Slope'].replace({ # same as above 
    'Up' : 0,
    'Flat':1,
    'Down':2
},inplace=True)
df['RestingECG'].replace({
    'Normal':0,
    'ST':1,
    'LVH':2  
},inplace=True)

# you could have just did [ df2 = pd.get_dummies(df,drop_first=True)
```

```python
df = pd.concat([df,dummies1,dummies2,dummies3],axis='columns')
df.head()
```

```python
df = df.drop(['Sex','ChestPainType','ExerciseAngina'],axis='columns')
```

```python
X = df.drop('HeartDisease',axis='columns')
y = df['HeartDisease']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
x_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=30)
```

```python
# (SVM, logistic regression, random forest) 
# and check which model gives you the best accuracy
# Now use PCA to reduce dimensions, retrain your model and see what impact it has on your model in terms of accuracy. 
# Keep in mind that many times doing PCA reduces the accuracy 
# but computation is much lighter and that's the trade off you need to consider while building models in real life
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
model_params = {
    'svm': {
        'model': SVC(gamma='auto'),
        'params': {
            'C' : [1,10,20],
            'kernel' : ['rbf','linear']
        }
    },
    'log': {
        'model':LogisticRegression(),
        'params':{
            'C':[1,5,10]
        }
    },
    'randomf': {
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    }
}
```

```python
scores = []

for model_name,mp in model_params.items():
    clf = GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(X_train,y_train)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
    
df = pd.DataFrame(scores)
df
```
